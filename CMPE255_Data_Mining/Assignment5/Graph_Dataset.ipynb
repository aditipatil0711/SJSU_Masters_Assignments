##Outlier Handling and Feature Engineering

### Outlier Handling
Given that we're dealing with citation data, and high citation counts are not necessarily errors but rather indications of highly influential work, we may not want to remove these outliers. However, for the purposes of certain analyses or machine learning models, we may want to reduce the impact of these extreme values.

We have a few options for handling outliers:

1. **Winsorizing**: Limiting extreme values to a certain percentile.
2. **Log Transformation**: Applying a log transformation to reduce the skewness of the distribution.
3. **Standardization**: While this does not affect the shape of the distribution, scaling the data to a mean of zero and a standard deviation of one can help with many machine learning algorithms.

### Feature Engineering
Feature engineering can involve creating new features or transforming existing ones to improve the model's performance. In the context of the citation data, we might consider:

1. **Citation Velocity**: The number of citations received in the last year compared to the total number of citations.
2. **Recent Impact Factor**: A ratio of citations since 2016 to the H-index since 2016.
3. **Citation Density**: The total number of citations divided by the H-index.

These new features could provide additional insights into the trends and impact of a researcher's work.


```python
import numpy as np


# Apply a log transformation to the citation count columns to reduce skewness
# We add 1 to the citation counts to avoid taking log of zero
citation_data_transformed = citation_data_cleaned.copy()
citation_count_columns = ['Citations 2020', 'Total_citation', 'Citation_since_2016']

# Apply log transformation and create new columns for the transformed values
for col in citation_count_columns:
    citation_data_transformed[f'log_{col}'] = np.log1p(citation_data_transformed[col])

# Now we will create new features
# Citation Velocity
citation_data_transformed['citation_velocity'] = citation_data_transformed['Citations 2020'] / citation_data_transformed['Total_citation']

# Recent Impact Factor
citation_data_transformed['recent_impact_factor'] = citation_data_transformed['Citation_since_2016'] / citation_data_transformed['H-index_since_2016']

# Citation Density
citation_data_transformed['citation_density'] = citation_data_transformed['Total_citation'] / citation_data_transformed['H-index']

# Display the head of the DataFrame to see the new columns
citation_data_transformed.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical columns (original and new features) for standardization
numerical_features = numerical_columns + [
    'log_Citations 2020', 'log_Total_citation', 'log_Citation_since_2016',
    'citation_velocity', 'recent_impact_factor', 'citation_density'
]

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the numerical features and transform
citation_data_standardized = citation_data_transformed.copy()
citation_data_standardized[numerical_features] = scaler.fit_transform(citation_data_transformed[numerical_features])

# Display the head of the DataFrame to see the standardized columns
citation_data_standardized[numerical_features].head()
```

## Clustering and Anomaly Detection

### Clustering

We'll use the K-means clustering algorithm, which is a popular method for partitioning the data into K distinct, non-overlapping subsets. To determine the optimal number of clusters (K), we can use the elbow method, which involves fitting the model with a range of K values and plotting the total within-cluster sum of square (WCSS) against the number of clusters. We look for a kink in the curve which indicates the elbow.

After clustering, we can label our data with the cluster IDs and analyze the characteristics of each cluster.

### Anomaly Detection

For anomaly detection, we can use the Isolation Forest algorithm, which is effective for high-dimensional datasets. It isolates anomalies instead of profiling normal data points, which is useful when anomalies are few and different.



```python
from sklearn.cluster import KMeans

# Determine the optimal number of clusters using the elbow method
wcss = []  # Within-cluster sum of square
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(citation_data_standardized[numerical_features])
    wcss.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Apply K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
cluster_labels = kmeans.fit_predict(citation_data_standardized[numerical_features])

# Add the cluster labels to the original DataFrame
citation_data_clustered = citation_data_cleaned.copy()
citation_data_clustered['Cluster'] = cluster_labels

# Check the distribution of data points among the clusters
cluster_distribution = citation_data_clustered['Cluster'].value_counts()

# Display the distribution and the first few rows of the DataFrame with cluster labels
cluster_distribution, citation_data_clustered.head()
```

```python
# Calculate the mean of the numerical features within each cluster
cluster_means = citation_data_clustered.groupby('Cluster')[numerical_columns].mean()

# Calculate the median of the numerical features within each cluster
cluster_medians = citation_data_clustered.groupby('Cluster')[numerical_columns].median()

cluster_means, cluster_medians
```

The cluster profiling based on the mean and median of the numerical features suggests the following characteristics for each cluster:

* Cluster 0: This cluster has the highest average and median H-index, citations for 2020, total citations, H-index since 2016, and citations since 2016. This indicates that Cluster 0 represents researchers with a high impact and a significant number of citations.
* Cluster 1: This cluster, with only 3 researchers, has the lowest average and median values for all citation metrics. These are likely to be researchers who are either newer to the field or have a lower citation impact.
* Cluster 2: This cluster represents the middle ground, with moderate average and median values for the citation metrics. Researchers in this cluster have a good number of citations but not as high as those in Cluster 0.


### Anomaly Detection with Isolation Forest

Now let's apply anomaly detection using the Isolation Forest algorithm. This will help us identify data points that are anomalies, which could be due to a variety of reasons such as data entry errors, true outliers in terms of impact, or other anomalies.

We will use the standardized numerical features for this analysis. Anomaly detection often requires a decision about the proportion of outliers we expect in the data; by default, we might start with 1% of the entries as anomalies unless we have a specific reason to choose a different threshold.

```python
from sklearn.ensemble import IsolationForest

# Initialize the IsolationForest
iso_forest = IsolationForest(n_estimators=100, max_samples='auto', contamination=float(.01),
                             random_state=42, verbose=0)

# Fit the model and predict anomalies
iso_forest.fit(citation_data_standardized[numerical_features])
anomalies = iso_forest.predict(citation_data_standardized[numerical_features])

# Anomalies are reported as -1 by the Isolation Forest algorithm
citation_data_clustered['Anomaly'] = anomalies
citation_data_clustered['Anomaly'] = citation_data_clustered['Anomaly'].apply(lambda x: 'Anomaly' if x == -1 else 'Normal')

# Count the number of anomalies detected
anomaly_counts = citation_data_clustered['Anomaly'].value_counts()

# Display the anomaly counts and the first few rows of the DataFrame with anomaly labels
anomaly_counts, citation_data_clustered.head()
```

```python
# Step 2 Actions:
# 1. Exclude identifier columns from the dataset
citation_data_features = citation_data_standardized.drop(columns=['Name', 'HomePage', 'Google_Scholar'])

# 2. Examine the number of unique values in 'Affilation' and 'Area of Research'
unique_values_affiliation = citation_data_features['Affilation'].nunique()
unique_values_area_of_research = citation_data_features['Area of Research'].nunique()

# Display the number of unique values
unique_values_affiliation, unique_values_area_of_research
```

## Data Splitting

With this feature selection, we can now prepare our final dataset for model training and evaluation. We will:

1. Split the data into a training set and a test set.
2. Use the training set to train the model and the test set to evaluate it.


```python
from sklearn.model_selection import train_test_split

# Final dataset with selected features
final_features = ['log_Citation_since_2016', 'log_Total_citation', 'citation_density', 'recent_impact_factor']
X_final = citation_data_model_ready[final_features]

# Since 'Cluster' is our target variable, we'll include it as well
y_final = citation_data_model_ready['Cluster']

# Split the dataset into a training set and a test set
# We'll use 80% of the data for training and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, random_state=42)

# Output the shape of the resulting datasets to confirm the split
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Building Models:

Building models manually before using AutoML can give us a baseline for comparison. Let's start with some common algorithms for classification tasks:

1. **Logistic Regression**: A good baseline for binary or multiclass classification problems.
2. **Random Forest**: An ensemble learning method for classification, regression, and other tasks.
3. **Gradient Boosting Machines (GBM)**: Another ensemble technique that produces a prediction model in the form of an ensemble of weak prediction models, typically decision trees.

For each model, we'll:

- Train the model on the training set.
- Evaluate the model on the test set using appropriate metrics (e.g., accuracy, precision, recall, F1-score).
- Compare the performance of the different models.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Initialize the Logistic Regression classifier
logistic_model = LogisticRegression(random_state=42)

# Fit the model to the training data
logistic_model.fit(X_train, y_train)

# Predict the labels for the test set
y_pred_logistic = logistic_model.predict(X_test)

# Evaluate the model's performance
logistic_accuracy = accuracy_score(y_test, y_pred_logistic)
logistic_classification_report = classification_report(y_test, y_pred_logistic)

logistic_accuracy, logistic_classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest classifier
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
random_forest_model.fit(X_train, y_train)

# Predict the labels for the test set
y_pred_rf = random_forest_model.predict(X_test)

# Evaluate the model's performance
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_classification_report = classification_report(y_test, y_pred_rf)

rf_accuracy, rf_classification_report
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize the Gradient Boosting classifier
gbm_model = GradientBoostingClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
gbm_model.fit(X_train, y_train)

# Predict the labels for the test set
y_pred_gbm = gbm_model.predict(X_test)

# Evaluate the model's performance
gbm_accuracy = accuracy_score(y_test, y_pred_gbm)
gbm_classification_report = classification_report(y_test, y_pred_gbm)

gbm_accuracy, gbm_classification_report
```

### Ensemble Model using Voting Classifier:


```python
from sklearn.ensemble import VotingClassifier

# Define the base classifiers
classifiers = [
    ('logistic', logistic_model),
    ('random_forest', random_forest_model),
    ('gbm', gbm_model)
]

# Initialize the Voting Classifier with soft voting
ensemble_model = VotingClassifier(estimators=classifiers, voting='soft')

# Fit the ensemble model to the training data
ensemble_model.fit(X_train, y_train)

# Predict the labels for the test set
y_pred_ensemble = ensemble_model.predict(X_test)

# Evaluate the ensemble model's performance
ensemble_accuracy = accuracy_score(y_test, y_pred_ensemble)
ensemble_classification_report = classification_report(y_test, y_pred_ensemble)

ensemble_accuracy, ensemble_classification_report
```
