```python
import numpy as np
import cv2

# Functions to calculate brightness and blur score
def calculate_brightness(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    return np.mean(img)

def calculate_blur_score(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    return cv2.Laplacian(img, cv2.CV_64F).var()

# Calculate brightness and blur scores for all images
brightness_scores = [calculate_brightness(img) for img in sample_image_paths]
blur_scores = [calculate_blur_score(img) for img in sample_image_paths]

# Convert to DataFrame for easier analysis
image_scores_df = pd.DataFrame({
    'Brightness': brightness_scores,
    'BlurScore': blur_scores
})

# Displaying summary statistics for brightness and blur scores
scores_summary = image_scores_df.describe()
scores_summary
```

### Color Distribution

```python
# Analyzing color distribution across sample images

def plot_color_distribution(image_path):
    # Load image using OpenCV (in BGR format)
    image = cv2.imread(image_path)
    # Convert BGR image to RGB
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Plot color distribution for each channel: R, G, and B
    colors = ('r', 'g', 'b')
    plt.figure(figsize=(10, 5))
    for i, color in enumerate(colors):
        histogram, bin_edges = np.histogram(
            image[:, :, i], bins=256, range=(0, 255)
        )
        plt.plot(bin_edges[0:-1], histogram, color=color)
    plt.title('Color Distribution')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    plt.legend(['Red', 'Green', 'Blue'])
    plt.show()

# Analyzing color distribution for a sample image
plot_color_distribution(sample_image_paths[0])
```


### Image Resolution or Quality Check:

For PNG images (like those in our dataset), the concept of "resolution" doesn't directly apply as it would with JPEG images, which have a quality factor. However, for PNGs, we can consider the level of detail or clarity in the image as an indicator of its quality.

A potential approach is to:
1. **Compute the entropy of an image**: Image entropy is a measure of the amount of information or randomness present in an image. A higher entropy value indicates more details, while a lower value might suggest potential blurriness or lack of detail.


```python
from skimage.measure import shannon_entropy

# Function to compute image entropy
def compute_entropy(image_path):
    # Load image using OpenCV (in grayscale for entropy calculation)
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    return shannon_entropy(image)

# Compute entropy for all sample images
entropy_values = [compute_entropy(img) for img in sample_image_paths]

# Convert to DataFrame for easier analysis
entropy_df = pd.DataFrame(entropy_values, columns=['Entropy'])

# Displaying summary statistics for entropy values
entropy_summary = entropy_df.describe()
entropy_summary
```

### Data Augmentation:

**Why Data Augmentation?**
Data augmentation is a strategy that enables practitioners to significantly increase the diversity of data available for training models, without actually collecting new data. Data augmentation techniques such as cropping, padding, and horizontal flipping are commonly used to train large neural networks.

**Steps in Data Augmentation**:

1. **Image Rotations**: Slight rotations of the image can simulate the variations of the position of the object.
2. **Horizontal & Vertical Flips**: These can help the model generalize better but should be used considering the data context.
3. **Random Cropping**: Taking a portion of the image randomly can simulate the effect of the object being partially visible.
4. **Brightness & Contrast Changes**: Modifying the brightness and contrast can simulate different lighting conditions.
5. **Zooming**: Zooming in and out can help simulate the object being closer or farther away.
6. **Shearing**: A form of distortion that can aid in generalizing the model.


```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

# Sample image for demonstration
sample_image_path = sample_image_paths[0]

# Initialize ImageDataGenerator with desired augmentations
datagen = ImageDataGenerator(
    rotation_range=30,  # Random rotations between 0 to 30 degrees
    width_shift_range=0.2,  # Horizontal shift
    height_shift_range=0.2,  # Vertical shift
    shear_range=0.2,  # Shearing
    zoom_range=0.2,  # Zooming
    horizontal_flip=True,  # Horizontal flip
    fill_mode='nearest',  # Filling strategy for created pixels
    brightness_range=(0.5, 1.5)  # Brightness changes
)

# Load the image
img = load_img(sample_image_path)
x = img_to_array(img)  # Convert image to numpy array
x = x.reshape((1,) + x.shape)  # Reshape for augmentation

# Generate augmented images and display them
plt.figure(figsize=(15, 10))
i = 0
for batch in datagen.flow(x, batch_size=1):
    plt.subplot(3, 3, i+1)
    plt.imshow(batch[0].astype('uint8'))
    plt.axis('off')
    i += 1
    if i == 9:  # Displaying 9 augmented versions
        break

plt.suptitle("Augmented Images")
plt.show()
```

```python
# Improving the augment_image function to handle different image modes

def augment_image_v2(image_path):
    # Load image using PIL
    original = Image.open(image_path).convert('RGB')  # Ensuring RGB mode

    # Rotation
    rotated = original.rotate(25)

    # Horizontal Flip
    flipped = original.transpose(Image.FLIP_LEFT_RIGHT)

    # Brightness Change
    from PIL import ImageEnhance
    enhancer = ImageEnhance.Brightness(original)
    brighter = enhancer.enhance(1.5)
    darker = enhancer.enhance(0.5)

    # Zooming (using cropping)
    width, height = original.size
    cropped = original.crop((0.1 * width, 0.1 * height, 0.9 * width, 0.9 * height))
    zoomed = cropped.resize((width, height))

    # Shearing using OpenCV (Affine Transformation)
    pts1 = np.float32([[5,5], [20,5], [5,20]])
    pts2 = np.float32([[1,10], [20,5], [10,25]])
    M = cv2.getAffineTransform(pts1,pts2)
    sheared = cv2.warpAffine(np.array(original), M, (width, height))

    return [original, rotated, flipped, brighter, darker, zoomed, Image.fromarray(sheared)]

# Apply augmentations and display
augmented_images = augment_image_v2(sample_image_path)

plt.figure(figsize=(15, 10))
for i, img in enumerate(augmented_images):
    plt.subplot(3, 3, i+1)
    plt.imshow(img)
    plt.title(image_titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

### Feature Extraction and Visualization

```python
from skimage.feature import hog
from sklearn.decomposition import PCA

# Extracting HOG features from a subset of images
def extract_hog_features(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    features, _ = hog(image, visualize=True, pixels_per_cell=(12, 12), cells_per_block=(2, 2), block_norm='L2-Hys')
    return features

# Extracting features for a subset of sample images (for demonstration and visualization)
hog_features = [extract_hog_features(img_path) for img_path in sample_image_paths[:200]]

# Applying PCA for dimensionality reduction (2 components for easy visualization)
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(hog_features)

# Visualizing the PCA-reduced features
plt.figure(figsize=(10, 8))
plt.scatter(reduced_features[:, 0], reduced_features[:, 1], alpha=0.6)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA-Reduced HOG Features')
plt.grid(True)
plt.show()
```

```python
# 1. Advanced Class Distribution Analysis:

# Getting the corresponding types for the sample images
sample_types = pokemon_df.loc[pokemon_df['Name'].isin([os.path.basename(img).split(".")[0] for img in sample_image_paths[:200]])]['Type1'].values

# Visualizing the PCA-reduced features with class labels
plt.figure(figsize=(12, 10))
for poke_type in pokemon_df['Type1'].unique():
    mask = (sample_types == poke_type)
    plt.scatter(reduced_features[mask, 0], reduced_features[mask, 1], label=poke_type, alpha=0.6)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA-Reduced HOG Features with Class Labels')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.show()
```

```python
from skimage.feature import greycomatrix, greycoprops

# 2. Texture Analysis:

# Function to extract texture features using GLCM
def extract_texture_features(image_path):
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    glcm = greycomatrix(image, [1], [0], 256, symmetric=True, normed=True)
    contrast = greycoprops(glcm, 'contrast')[0, 0]
    dissimilarity = greycoprops(glcm, 'dissimilarity')[0, 0]
    homogeneity = greycoprops(glcm, 'homogeneity')[0, 0]
    energy = greycoprops(glcm, 'energy')[0, 0]
    correlation = greycoprops(glcm, 'correlation')[0, 0]
    return contrast, dissimilarity, homogeneity, energy, correlation

# Extracting texture features for a subset of sample images
texture_features = [extract_texture_features(img_path) for img_path in sample_image_paths[:200]]
texture_df = pd.DataFrame(texture_features, columns=['Contrast', 'Dissimilarity', 'Homogeneity', 'Energy', 'Correlation'])

# Displaying summary statistics for texture features
texture_summary = texture_df.describe()
texture_summary
```

### Model Training:



```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Encoding the 'Type1' labels
le = LabelEncoder()
encoded_labels = le.fit_transform(sample_types)

# Standardizing the extracted features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(extracted_features)

# Splitting the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(scaled_features, encoded_labels, test_size=0.2, stratify=encoded_labels, random_state=42)

# Using Logistic Regression as a simple classifier
clf = LogisticRegression(max_iter=1000, random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Calculating accuracy for baseline
baseline_accuracy = accuracy_score(y_test, y_pred)

baseline_accuracy
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# 1. Select Models:
# Instantiating the models
svm_model = SVC(kernel='linear', random_state=42)
rf_model = RandomForestClassifier(random_state=42)

# 2. Train Models:
svm_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)

# 3. Evaluate Models on Validation Data:
svm_predictions = svm_model.predict(X_test)
rf_predictions = rf_model.predict(X_test)

# Classification reports
svm_report = classification_report(y_test, svm_predictions)
rf_report = classification_report(y_test, rf_predictions)

svm_report, rf_report
```
