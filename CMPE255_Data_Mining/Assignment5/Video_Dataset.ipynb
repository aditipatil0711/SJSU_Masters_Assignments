```python
import zipfile
import os

# Unzipping the dataset
zip_path = 'video-dataset.zip'
extract_folder = 'video-dataset'

# Ensure the extract folder is created
os.makedirs(extract_folder, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

# List the files in the extracted folder
extracted_files = os.listdir(extract_folder)
extracted_files
```

```python
# Function to explore the contents of the folders within the dataset
def explore_folder(folder_path):
    folder_contents = os.listdir(folder_path)
    file_types = set()
    num_files = len(folder_contents)
    example_file = None

    # Identify file types and get an example file
    for file in folder_contents:
        file_type = file.split('.')[-1]
        file_types.add(file_type)
        if not example_file:
            example_file = file

    return {
        'number_of_files': num_files,
        'file_types': file_types,
        'example_file': example_file,
        'all_files': folder_contents
    }

# Exploring the 'animals' folder
animals_folder_path = os.path.join(extract_folder, 'animals')
animals_folder_info = explore_folder(animals_folder_path)
animals_folder_info
```

```python
import cv2

# Function to extract metadata from a video file
def extract_video_metadata(video_path):
    # Using OpenCV to capture the video
    video_capture = cv2.VideoCapture(video_path)

    # Extracting metadata
    metadata = {
        'width': int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH)),
        'height': int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT)),
        'fps': video_capture.get(cv2.CAP_PROP_FPS),
        'frame_count': int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT)),
    }

    # Release the video capture object
    video_capture.release()

    return metadata

# Path to the example video file
example_video_path = os.path.join(animals_folder_path, animals_folder_info['example_file'])

# Extracting metadata from the example video file
video_metadata = extract_video_metadata(example_video_path)
video_metadata
```

```python
# Calculating the video duration using the frame count and frames per second (fps)
video_duration_seconds = video_metadata['frame_count'] / video_metadata['fps']
video_duration_minutes = video_duration_seconds / 60  # Convert to minutes

# Storing the duration in the metadata dictionary
video_metadata['duration_seconds'] = video_duration_seconds
video_metadata['duration_minutes'] = video_duration_minutes

# To generate a thumbnail, we'll capture a frame from the middle of the video
def capture_thumbnail(video_path, frame_number):
    video_capture = cv2.VideoCapture(video_path)
    # Set the frame position
    video_capture.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    success, frame = video_capture.read()
    if success:
        # Save the frame as a thumbnail
        thumbnail_path = video_path.replace('.mp4', '_thumbnail.jpg')
        cv2.imwrite(thumbnail_path, frame)
    # Release the video capture object
    video_capture.release()
    return thumbnail_path if success else None

# Capture a thumbnail from the middle of the video
middle_frame_number = video_metadata['frame_count'] // 2
thumbnail_path = capture_thumbnail(example_video_path, middle_frame_number)

# Update the metadata with the thumbnail path
video_metadata['thumbnail_path'] = thumbnail_path
video_metadata, thumbnail_path
```

### Step 4: Data Cleaning

1. **Missing or Corrupted Files**: We'll check to ensure all video files are accessible and not corrupted.
2. **Consistency Across Videos**: We'll verify that all videos within the same category have similar properties (such as resolution and frame rate) to ensure consistency.

```python
import pandas as pd

# Function to check video files and extract metadata
def check_and_extract_metadata(folder_path):
    video_files = os.listdir(folder_path)
    metadata_list = []
    corrupted_files = []

    for video_file in video_files:
        video_path = os.path.join(folder_path, video_file)
        try:
            metadata = extract_video_metadata(video_path)
            metadata['file_name'] = video_file
            metadata_list.append(metadata)
        except Exception as e:
            corrupted_files.append(video_file)

    return pd.DataFrame(metadata_list), corrupted_files

# Check and extract metadata for each category
animals_metadata_df, animals_corrupted = check_and_extract_metadata(animals_folder_path)
food_metadata_df, food_corrupted = check_and_extract_metadata(food_folder_path)
insects_metadata_df, insects_corrupted = check_and_extract_metadata(insects_folder_path)

# Combine the metadata into a single DataFrame
all_videos_metadata = pd.concat([animals_metadata_df, food_metadata_df, insects_metadata_df], ignore_index=True)

# Display the combined metadata and any corrupted files
all_videos_metadata, {'animals_corrupted': animals_corrupted, 'food_corrupted': food_corrupted, 'insects_corrupted': insects_corrupted}
```

```python
# Cleaning the metadata by excluding thumbnail images
# Thumbnails have a frame count of 1 and their file names contain 'thumbnail'
videos_metadata_cleaned = all_videos_metadata[~all_videos_metadata['file_name'].str.contains('thumbnail')]
videos_metadata_cleaned = videos_metadata_cleaned[videos_metadata_cleaned['frame_count'] > 1]

# Resetting the index after cleaning
videos_metadata_cleaned.reset_index(drop=True, inplace=True)

# We'll also add a column for video duration in seconds for further analysis
videos_metadata_cleaned['duration_seconds'] = videos_metadata_cleaned['frame_count'] / videos_metadata_cleaned['fps']

# Now we will visualize the relationships in the metadata
# Specifically, we want to plot:
# 1. Resolution (width x height) vs. Duration
# 2. Frame Count vs. Duration

# Creating a new column for resolution area (width x height)
videos_metadata_cleaned['resolution_area'] = videos_metadata_cleaned['width'] * videos_metadata_cleaned['height']

# Creating the plots
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the aesthetic style of the plots
sns.set_style("whitegrid")

# Plot 1: Resolution Area vs. Duration
plt.figure(figsize=(14, 7))
sns.scatterplot(data=videos_metadata_cleaned, x='resolution_area', y='duration_seconds', hue='file_name', s=100)
plt.title('Resolution Area vs. Duration')
plt.xlabel('Resolution Area (width x height)')
plt.ylabel('Duration (seconds)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()

# Plot 2: Frame Count vs. Duration
plt.figure(figsize=(14, 7))
sns.scatterplot(data=videos_metadata_cleaned, x='frame_count', y='duration_seconds', hue='file_name', s=100)
plt.title('Frame Count vs. Duration')
plt.xlabel('Frame Count')
plt.ylabel('Duration (seconds)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.show()
```

## Anomaly Detection:

Anomaly detection in the context of this video dataset could involve identifying videos that have unusual properties compared to the rest of the dataset. These could be videos with an exceptionally high or low frame count, resolution, duration, or any other significant deviations in their metadata.

Here's how we can approach anomaly detection:

1. **Statistical Analysis**: Use statistical methods to identify outliers. For example, we could use the Interquartile Range (IQR) method or Z-scores to find anomalies in numerical metadata such as duration, frame count, and resolution.
2. **Visualization**: Box plots can be useful to visually spot outliers in the dataset.
3. **Cluster-Based Detection**: If we consider the multidimensional nature of video metadata (e.g., duration, resolution, frame count), clustering algorithms like K-Means or DBSCAN could help identify groups of similar videos and spot any videos that do not belong to any cluster (potential anomalies).


```python
# Function to calculate IQR and detect outliers
def detect_outliers_iqr(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detecting outliers for 'duration_seconds', 'frame_count', and 'resolution_area'
duration_outliers, duration_lower, duration_upper = detect_outliers_iqr(videos_metadata_cleaned, 'duration_seconds')
frame_count_outliers, frame_count_lower, frame_count_upper = detect_outliers_iqr(videos_metadata_cleaned, 'frame_count')
resolution_area_outliers, resolution_area_lower, resolution_area_upper = detect_outliers_iqr(videos_metadata_cleaned, 'resolution_area')

# Displaying outliers for each category
outliers_summary = {
    'Duration Outliers': duration_outliers[['file_name', 'duration_seconds']],
    'Frame Count Outliers': frame_count_outliers[['file_name', 'frame_count']],
    'Resolution Area Outliers': resolution_area_outliers[['file_name', 'resolution_area']]
}

outliers_summary
```
