# Step:3 Initiatialize Pycaret and work!

Here , I have imported the classification experiment library from pycaret. The setup() command helps initialize the setup

```python
from pycaret.classification import *
X = data.drop('Outcome', axis=1)
y = data['Outcome']
clf = setup(data = data, target = 'Outcome',train_size=0.8,numeric_imputation='mean',categorical_imputation='mode',feature_selection=True)
```

# Compare Models

This function uses cross-validation to train and assess the performance of every estimator in the model library. It suggests the best output providing model

```python
best_model = compare_models(n_select=5)
```

```python
linear_discriminant_analysis = create_model('lda',fold =10)
```

## Tune Model

```python
tuned_linear_discriminant_analysis= tune_model(linear_discriminant_analysis)
```

# Evaluate Model
Evaluate Model is a method of plotting data. An Alternative to plot_model function



```python
evaluate_model(tuned_linear_discriminant_analysis)
```

# Model Prediction:

The predict_model function adds new columns to the dataframe: prediction_label and prediction_score, which represent the predicted class and its probability, respectively. If no data is provided (default setting), it scores using the test set established by the setup function.

```python
predict_model(tuned_linear_discriminant_analysis)
```

```python
final_model = finalize_model(tuned_linear_discriminant_analysis)
```

```python
new_data = data.copy().drop('Outcome', axis=1)
new_data.head()
```

```python
predictions = predict_model(final_model, data = new_data)
predictions.head()
```
