# Step:3 Initiatialize Pycaret and work!

Here , I have imported the classification experiment library from pycaret. The setup() command helps initialize the setup

```python
from pycaret.clustering import *
data = setup(data,session_id=123,use_gpu=True)
```

# Create Model

This function trains a specified model and assesses its performance. To view the evaluated metrics, use the `get_metrics` function. You can introduce or delete custom metrics with the `add_metric` and `remove_metric` functions, respectively. To see all potential models, use the `models` function.

```python
kmeans = create_model('kmeans')
```

```python
models()
```

## Assign Model
This function assigns cluster labels to the training data, given a trained model.

```python
results = assign_model(kmeans)
results.head()
```

```python
evaluate_model(kmeans)
```

# Analyze Model

The plot_model function allows you to visualize the performance of a trained model on the test data.


```python
plot_model(kmeans,plot='cluster')
```

# Model Prediction:

The predict_model function adds new columns to the dataframe: prediction_label and prediction_score, which represent the predicted class and its probability, respectively. If no data is provided (default setting), it scores using the test set established by the setup function.

```python
new_data = pd.read_csv('housing.csv')
new_data.head()
```

```python
predictions = predict_model(kmeans,data=new_data)
predictions.head()
```

# Saving a Model

We can save this model for later use as well with this function

```python
save_model(kmeans,'kmeans_model')
```
