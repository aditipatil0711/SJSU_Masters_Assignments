# Step:3 Initiatialize Pycaret and work!

Here , I have imported the classification experiment library from pycaret. The setup() command helps initialize the setup

```python
from sklearn.model_selection import train_test_split
data_frame = data.drop(data,axis=1)
df_tr, df_te = train_test_split(data_frame, test_size=0.2, random_state=42)
```

```python
from pycaret.regression import *
data = setup(data, target = 'charges', session_id = 123,use_gpu=True)
```

# Compare Models

This function uses cross-validation to train and assess the performance of every estimator in the model library. It suggests the best output providing model

```python
best = compare_models()
```

# Evaluate Model
Evaluate Model is a method of plotting data. An Alternative to plot_model function


```python
evaluate_model(best)
```

# Model Prediction:

The predict_model function adds new columns to the dataframe: prediction_label and prediction_score, which represent the predicted class and its probability, respectively. If no data is provided (default setting), it scores using the test set established by the setup function.

```python
pred_holdout = predict_model(best)
```

```python
new_data = pd.read_csv('insurance.csv')
new_data.head()
```

```python
new_data.drop('charges',axis=1)
predictions = predict_model(best, data=new_data)
```

# Saving a Model
