# **G. Timeseries Forcasting : Univariate without exogenous variables  Using Pycaret**

Univariate analysis focuses on examining a single variable, typically through summary statistics or visualization. In the context of time series, "univariate without exogenous variables" means predicting future values based solely on past values of the same series. No external factors or variables are considered in the analysis or prediction.

In this example, we will try to forecast electic production based on timeseries data

```python
data = pd.read_csv('Electric_Production.csv')
data.rename(columns={'DATE':'date','IPG2211A2N':'production'},inplace=True)
data.head()
#from pycaret.datasets import get_data
#data = get_data('airline')
```

```python
data['date'] = pd.to_datetime(data['date'])
data.columns
```

# Step:3 Initiatialize Pycaret and work!

Here , I have imported the classification experiment library from pycaret. The setup() command helps initialize the setup

```python
from pycaret.time_series import *
data_ts = setup(data = data, target='production',fh = 3)

```

# Compare Models

```python
best  = compare_models()
```

# Analyze Model

The plot_model function allows you to visualize the performance of a trained model on the test data.


```python
plot_model(best, plot = 'forecast')
```

# Model Prediction:

The predict_model function adds new columns to the dataframe: prediction_label and prediction_score, which represent the predicted class and its probability, respectively. If no data is provided (default setting), it scores using the test set established by the setup function.

```python
pred_holdout = predict_model(best)
```

```python
prediction = predict_model(best, fh = 6)
prediction
```

# Saving a Model

We can save this model for later use as well with this function

```python
save_model(best, 'best_model')
```
