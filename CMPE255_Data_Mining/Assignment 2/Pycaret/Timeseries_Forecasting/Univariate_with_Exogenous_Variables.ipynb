# **G. Timeseries Forcasting : Univariate with exogenous variables  Using Pycaret**

Univariate analysis with exogenous variables involves studying a single target variable while incorporating external (exogenous) variables to enhance predictions or understanding. In time series, this means forecasting future values of a series based on its past values and additional external factors. These exogenous variables provide context and can improve prediction accuracy.

In this example, we will try to forecast air quality based on timeseries data

```python
data = pd.read_csv('airquality.csv')
data.head()
```

# Preprocessing the Data

Changing format of datetime so it can be the index.

```python
import datetime
data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])
data.drop(columns=["Date", "Time"], inplace=True)
data.replace(-200, np.nan, inplace=True)
data.set_index("datetime", inplace=True)
data.head()
```

```python
metric = 'rmse'
exclude =  ["auto_arima", "bats", "tbats", "lar_cds_dt", "par_cds_dt"]
```

# Step:3 Initiatialize Pycaret and work!

Here , I have imported the classification experiment library from pycaret. The setup() command helps initialize the setup

```python
from pycaret.time_series import *
data_ts = setup(data, target = 'CO(GT)', fh = 12,session_id=123,enforce_exogenous=False,
    numeric_imputation_target="ffill", numeric_imputation_exogenous="ffill",use_gpu = True)
```

# Compare Models

```python
best_model = compare_models(exclude=exclude)
```

# Analyze Model

The plot_model function allows you to visualize the performance of a trained model on the test data.


```python
plot_model(best_model, plot = 'forecast')
```

# Model Prediction:

The predict_model function adds new columns to the dataframe: prediction_label and prediction_score, which represent the predicted class and its probability, respectively. If no data is provided (default setting), it scores using the test set established by the setup function.

```python
pred_holdout = predict_model(best_model)
```

```python
predictions = predict_model(best_model, fh = 36)
predictions.head()
```
